```python
import pandas as pd
full_df=pd.read_csv(r"C:\Users\anany\OneDrive\Desktop\learner_analytics\data\processed\edupro_clean_data.csv")
```

```python
total_enrollments = full_df.shape[0]
```

```python
enrollments_by_age = (
    full_df
    .groupby("AgeGroup")["CourseID"]
    .count()
    .reset_index(name="Enrollments")
)
```

```python
gender_participation = (
    full_df["Gender"]
    .value_counts(normalize=True)
    .reset_index()
)

gender_participation.columns = ["Gender", "ParticipationRatio"]
gender_participation["ParticipationRatio"] *= 100
```

```python
category_popularity = (
    full_df
    .groupby("CourseCategory")["CourseID"]
    .count()
    .reset_index(name="Enrollments")
    .sort_values(by="Enrollments", ascending=False)
)
```

```python
level_distribution = (
    full_df["CourseLevel"]
    .value_counts(normalize=True)
    .reset_index()
)

level_distribution.columns = ["CourseLevel", "Percentage"]
level_distribution["Percentage"] *= 100
```
